```python
docs=[]
for link in links:
    docs.append({"Filing/Report Type":link.get_text(),"href":link.get('href').replace("\\","/")})
df=pd.DataFrame(docs)
df
```

```python
traveldocs=df.drop([0,1,2,1137,1138]).reset_index()
```

```python
traveldocs
```

```python
fulltable=traveltable
```

```python
fulltable["href"]=traveldocs["href"]
```

```python
fulltable
```

```python
leftside='http://giftrule-disclosure.senate.gov'
```

```python
for i in range(0, len(fulltable)):
    row=fulltable.iloc[i]
    url=leftside+row["href"]
    name=f"Senate Gift Travel Docs/{row['Last Name']}{row['Reporting Year']}.pdf"
    print(f"now downloading #{i} URL {url}")
    wget.download(url, out=name)
    time.sleep(1)
print("Download complete")
```
